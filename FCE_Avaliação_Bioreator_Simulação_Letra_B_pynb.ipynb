```python
from DisciplinaFCE.FCE_Bioreator import(
    ParametrosBiorreator,
    Vmax_Arrhenius,
    Biorreator_modelo,
    Simula_Biorreator,
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros do modelo
params = ParametrosBiorreator()

# Condição inicial
S0 = 100.0  # mg/L

# Tempo total da simulação
tempo_final = 30.0  # h

# Temperaturas em °C
temperaturas = [25, 35, 45]

# Dicionário para armazenar resultados finais
resultados_finais = {}

# Figura
plt.figure(figsize=(8, 5))

# Simulações
for T in temperaturas:

    tempo, concentracao = Simula_Biorreator(
        S0,
        tempo_final,
        T,
        params
    )

    resultados_finais[T] = concentracao[-1]

    plt.plot(
        tempo,
        concentracao,
        label=f"{T} °C"
    )

# Configuração do gráfico
plt.title("Consumo de substrato em diferentes temperaturas")
plt.xlabel("Tempo (h)")
plt.ylabel("Concentração de substrato (mg/L)")
plt.legend()
plt.grid(True)

plt.show()

# Resultados finais
print("\nConcentração final de substrato:")

for T, S_final in resultados_finais.items():
    print(f"{T} °C -> {S_final:.4f} mg/L")
```

# Explicação Teórica

Observa-se que o aumento da temperatura acelera o consumo de substrato no biorreator. A 45 °C a concentração de substrato diminui mais rapidamente, pois o valor de Vmax aumenta devido à dependência da temperatura pela equação de Arrhenius.

Já a 25 °C o consumo ocorre de forma mais lenta, resultando em uma concentração final maior ao final da simulação.Portanto, temperaturas mais elevadas favorecem a velocidade de reação e a degradação do substrato.
